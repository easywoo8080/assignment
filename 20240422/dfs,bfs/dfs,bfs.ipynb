### ğŸ’¡ í(Queue - FIFO : First In, First Out)
- íëŠ” ë°ì´í„°ë¥¼ ì²˜ìŒì— ë„£ì€ ê²ƒë¶€í„° ì°¨ë¡€ëŒ€ë¡œ êº¼ë‚´ëŠ” ì„ ì…ì„ ì¶œì˜ ìë£Œêµ¬ì¡°ì´ë‹¤.
- ì¦‰, ì²˜ìŒ ì‚½ì…ëœ ìš”ì†Œê°€ ê°€ì¥ ë¨¼ì € ë‚˜ì˜¤ëŠ” êµ¬ì¡°ì´ë‹¤.
- pythonì˜ ê²½ìš° ë¦¬ìŠ¤íŠ¸ë¥¼ ì‚¬ìš©í•´ë„ ë˜ê³  collections ëª¨ë“ˆì˜ `deque`ë¥¼ ì‚¬ìš©í•´ë„ ë˜ì§€ë§Œ `popleft`ê°€ ìˆì–´ ì‹œê°„ ë³µì¡ë„ê°€ ë” ë‚®ì€ `deque`ë¥¼ ë” ë§ì´ ì‚¬ìš©í•œë‹¤.

### íë¥¼ í™œìš©í•œ BFSì˜ˆì‹œ

<img src="img/ex1.png" width="70%">

```python
from collections import deque

# BFS ë©”ì„œë“œ ì •ì˜
def bfs(graph, start, visited):
    # í(Queue) êµ¬í˜„ì„ ìœ„í•´ deque ë¼ì´ë¸ŒëŸ¬ë¦¬ ì‚¬ìš©
    queue = deque([start])
    # í˜„ì¬ ë…¸ë“œë¥¼ ë°©ë¬¸ ì²˜ë¦¬
    visited[start] = True
    # íê°€ ë¹Œ ë•Œê¹Œì§€ ë°˜ë³µ
    while queue:
        # íì—ì„œ í•˜ë‚˜ì˜ ì›ì†Œë¥¼ ë½‘ì•„ ì¶œë ¥í•˜ê¸°
        v = queue.popleft()
        print(v, end=' ')
        # ì•„ì§ ë°©ë¬¸í•˜ì§€ ì•Šì€ ì¸ì ‘í•œ ì›ì†Œë“¤ì„ íì— ì‚½ì…
        for i in graph[v]:
            if not visited[i]:
                queue.append(i)
                visited[i] = True
# ê° ë…¸ë“œê°€ ì—°ê²°ëœ ì •ë³´ë¥¼ í‘œí˜„ (2ì°¨ì› ë¦¬ìŠ¤íŠ¸)
graph = [
    [], # ë³´í†µ ê·¸ë˜í”„ê°€ 1ë²ˆë¶€í„° ì‹œì‘í•˜ëŠ” ê²½ìš°ê°€ ë§ì•„ì„œ 0ë²ˆì§¸ëŠ” ë¹„ì›Œë‘ 
    [7, 5, 3], # 1ë²ˆ ë…¸ë“œì™€ ì¸ì ‘í•œ ë…¸ë“œë“¤
    [],        # 2ë²ˆ ë…¸ë“œì™€ ì¸ì ‘í•œ ë…¸ë“œë“¤
    [6],
    [9], 
    [8, 10], 
    [], 
    [4], 
    [2],
    [], 
    []
]
# ê° ë…¸ë“œê°€ ë°©ë¬¸ëœ ì •ë³´ë¥¼ í‘œí˜„ (1ì°¨ì› ë¦¬ìŠ¤íŠ¸)
visited = [False] * 11

bfs(graph, 1, visited)
```
